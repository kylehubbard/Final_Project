```python
# Start final project import modules
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
%matplotlib inline
import sys
import string
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder 
from sklearn import feature_extraction
from __future__ import print_function
from sklearn.ensemble import RandomForestClassifier
```

```python
# read in the datasets
 
jd = pd.read_csv('jobs_data.csv')
```

```python
#1. Explore the dataset: 
    # There is a mix of string and numeric data. There are some variables that can be converted to dummy or binary values.
    # While there are no NAs there are unknown values in some of the string data which may need to be treated as NAs.

jd.describe(include='all')
jd.info()
```

```python
#Input variables:
# Job.Title.og  - customer entered job title
# jobs - cleaned customer entered job title, formatted to remove punctuation, make lower case, fix spelling errors, abbreviations etc.
# Industry - clustered NAICS codes
# CBSA.Title - Census Bureau Statistical Area matched by city, state
# STATE - client state
# an.sal - employee annual salary
# bonus - employee bonus
# Age - employee age
# BLS.Verticals - industries designated by BLS NAICS codes
# CITY - client city
# POSTAL - client zip code
# city.state - client city and state
# EG.abbr.1 - employee ethnicity
# Full.Time.Part.Time - employee full time or part time
# Gender - Male, Female, unknown
# Marital.Status - Married, Single, unknown
# NAICS.2 - 2 digit NAICS code
# NAICS.2.des - 2 digit NAICS code description
# NAICS.3 - 3 digit NAICS code
# NAICS.3.des - 3 digit NAICS code description
# NAICS.4 - 4 digit NAICS code
# NAICS.4.des - 4 digit NAICS code description
# NAICS.5 - 5 digit NAICS code
# NAICS.5.des - 5 digit NAICS code description
# Regular.Temporary - employee regular or temporary
# Salaried.Hourly - employee salaried or hourly
# Workers.Comp.Code - workers comp code
# Job.Code - EEOC reporting code
# Job.Code.Description - EEOC reporting code description
# supervisor - is empoloyee a supervisor 0/1
# Collapsing.Job.Title - target job titles
# groups - Functional job groupings: Admin, Executive, Marketing, Operations, Sales, Software Engineering
# uid - unique observation identifier
```

```python
# Select variables to encode

# There are many variables that can be excluded due to similarity with other columns, but should be used to test predictive power

catagorical_var = ['jobs', 'Industry', 'city.state', 'EG.abbr.1', 'Gender', 'Marital.Status','Workers.Comp.Code',
                   'Job.Code', 'groups', 'Collapsing.Job.Title']

binary_var = ['Full.Time.Part.Time', 'Regular.Temporary', 'Salaried.Hourly']

numeric_var = ['an.sal', 'bonus', 'Age', 'supervisor']
    
string_variables = catagorical_var + binary_var

cat_columns = string_variables + numeric_var

jd_1 = jd[cat_columns]
#jd_1.head()
```

```python
from sklearn import preprocessing

df_numerical = jd_1.copy()
encoder_dict = {}

for col in string_variables:
    le = preprocessing.LabelEncoder()
    le.fit(jd_1[col])
    encoder_dict[col] = le
    df_numerical[col] = le.transform(jd_1[col])

df_numerical.head()
```

```python
feature_cols = list(df_numerical.columns.values)

feature_cols.remove('Collapsing.Job.Title')
len(feature_cols)
```

```python

# define X and y
X = df_numerical[feature_cols].values
y =df_numerical['Collapsing.Job.Title'].values

# split into train/test
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python

treeclf = RandomForestClassifier(random_state=1)
treeclf.fit(X_train, y_train)
```

```python
pd.DataFrame({'feature':feature_cols, 'importance':treeclf.feature_importances_}).sort('importance', ascending = False)
```

```python
# 4. Evaluate the accuracy of the decision tree. 
#scores = cross_val_score(treeclf, X, y, cv = 5, scoring = 'accuracy')
scores = cross_val_score(treeclf, X_train, y_train, cv = 5, scoring = 'accuracy')
cvs_1 = np.mean(scores)
cvs_1
```

```python
clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```
